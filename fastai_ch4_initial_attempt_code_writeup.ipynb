```
# This is the code my initial attempt to structure the data for traning. My
# plan was to make a list contain dictionaries that had keys for the number 
# labels and vaues for the pixel values. I have not included the step to 
# reorganise the  array to a 28x28 np array as I was not sure this is step woud
# be necessary for the calculations.

# However I ultimately aboanded this approach as interating over al 6000 entries
# cannot be done performantly. I suspect if I coud reorganise data into labelled
# arrays or tensor rather than looping and generating lists the preformance may
# improve

mnist_images = open('/content/drive/My Drive/datasets/mnist_train.csv')
mnist_images_df = pd.read_csv(mnist_images)

# Function to generate number label and array of pixel vals from df file
# x = df, y = slice for label and pixel values

def gen_label_pixel_vals(x,y):

  # Dictionary to hold number label and pixel vals associated with 
  # number

  number_label_pixels = dict()
  
  # Generate a series containing label and pixel val numbers

  number_series = x.loc[y]

  # Generate label for number

  number_label = number_series[0]
  
  # Generate series of pixel values

  number_pixel_vals_series = number_series[1:]

  # Convert series to array

  number_pixel_vals_array = np.array(number_pixel_vals_series)

  # Assigning label and pixel array to number_label_pixels dict

  number_label_pixels[number_label] = number_pixel_vals_array

  return number_label_pixels


# Function to generate list of dicts containing their number labels and pixel
# values

# x is a DF containing number labels and pixel values, z is range of entries to
# be evaluated. The sample data has 6000 examples but I will limit that to 10
# for example 

def gen_complete_label_pixels(x,z):
    
    # empty list for number labels and pixel values

    complete_labels_pixels = []

    for i in range(z):
      complete_labels_pixels.append(gen_label_pixel_vals(x,i))

    return complete_labels_pixels



gen_complete_label_pixels(mnist_images_df, 10)
```
